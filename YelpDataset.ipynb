# Delving into user dataset first

In this section, we explore the **user dataset** to evaluate whether it is suitable for building a recommendation system.  

### 🔹 Activity Distribution
- **Goal**: Do most users write many reviews, or just 1–2? (data sparsity problem)
- **Findings**:  
  - The distribution of reviews per user is highly skewed.  
  - Most users only write a handful of reviews (1–5).  
  - A small fraction of “power users” contribute thousands of reviews.  
- **Implication**:  
  - Data sparsity is a major challenge for collaborative filtering methods.  
  - Cold-start users (those with very few reviews) will require alternative strategies (e.g., content-based features or hybrid models).

### 🔹 Social Features
- **Goal**: Do users have friends, fans, and elite status? Could these be used in social-based recommenders?
- **Findings**:  
  - Many users have **zero friends**, meaning the social graph is sparse.  
  - Some users, however, do have large friend networks.  
  - A small number of users are marked as “elite,” representing loyal/power users.  
- **Implication**:  
  - Social recommenders could be attempted, but sparsity limits their effectiveness.  
  - Elite users might serve as an interesting segment to study separately.

### 🔹 Rating Behavior
- **Goal**: Explore the average star rating per user to see whether users are biased (e.g., always positive or always negative).  
- **Findings**:  
  - Most users cluster around 3–4 stars.  
  - There is limited variance in rating behavior.  
- **Implication**:  
  - Low variance in ratings can reduce the ability of a recommender system to differentiate between businesses.  
  - Textual reviews may need to be leveraged to enrich the signal.

### 🔹 Fans & Compliments
- **Goal**: Investigate extra signals about user style (fans, compliments).  
- **Findings**:  
  - These features exist but are very sparse.  
  - Only a small subset of users accumulate significant fans or compliments.  
- **Implication**:  
  - While interesting for advanced models, these features are unlikely to provide broad coverage for a baseline recommender.
