```python
import json

global_path = r"C:\Users\lebro\OneDrive - Nanyang Technological University\Github\YelpFYP\step4_outputs\summary_global.json"
mobility_path = r"C:\Users\lebro\OneDrive - Nanyang Technological University\Github\YelpFYP\step4_outputs\summary_mobility_aware.json"
lightfm_path = r"C:\Users\lebro\OneDrive - Nanyang Technological University\Github\YelpFYP\lightfm_outputs\lightfm_metrics.json"

with open(global_path, "r", encoding="utf-8") as f:
    global_summary = json.load(f)

with open(mobility_path, "r", encoding="utf-8") as f:
    mobility_summary = json.load(f)

with open(lightfm_path, "r", encoding="utf-8") as f:
    lightfm_summary = json.load(f)

print(global_summary)
print(mobility_summary)
print(lightfm_summary)
```

```python
global_row = {
    "model": "global_popularity",
    **global_summary
}

mobility_row = {
    "model": "mobility_aware",
    **mobility_summary
}

lightfm_row = {
    "model": "LightFM-WARP",
    "n_users_evaluated": lightfm_summary["num_eval_users"],
    "hit_rate": lightfm_summary["hit_rate_at_100"],   # fair match to Step 4 hit_rate
    "Recall@10": lightfm_summary["recall_at_10"],
    "Recall@20": lightfm_summary["recall_at_20"],
    "MRR": lightfm_summary["mrr"],
    "NDCG@10": lightfm_summary["ndcg_at_10"],
    "NDCG@20": lightfm_summary["ndcg_at_20"],
}
```

```python
comparison_table = pd.DataFrame([
    global_row,
    mobility_row,
    lightfm_row
])

comparison_table
```

```python
metric_cols = ["hit_rate", "Recall@10", "Recall@20", "MRR", "NDCG@10", "NDCG@20"]

comparison_table[metric_cols] = comparison_table[metric_cols].round(4)

comparison_table
```

```python
comparison_table.to_csv("report_model_comparison.csv", index=False)
print("Saved report_model_comparison.csv")
```
