# 5-core filtering

Replicating Section 4.2.1

```python
MIN_REVIEWS = 5

# Filter users
user_counts = core["user_id"].value_counts()
valid_users = user_counts[user_counts >= MIN_REVIEWS].index

# Filter businesses
biz_counts = core["business_id"].value_counts()
valid_biz = biz_counts[biz_counts >= MIN_REVIEWS].index

core_filtered = core[
    core["user_id"].isin(valid_users) &
    core["business_id"].isin(valid_biz)
]

print("Before filtering:", len(core))
print("After 5-core filtering:", len(core_filtered))
```

```python
core_filtered["date"] = pd.to_datetime(core_filtered["date"])

core_filtered = core_filtered.sort_values(
    by=["user_id", "date"]
).reset_index(drop=True)
```

# Geospatial feature engineering

Replicating Section 3.1.1

```python
def compute_gen_bbox(df):
    df = df.copy()
    df["min_lat"] = df["latitude"].expanding().min()
    df["max_lat"] = df["latitude"].expanding().max()
    df["min_lon"] = df["longitude"].expanding().min()
    df["max_lon"] = df["longitude"].expanding().max()
    return df

core_filtered = core_filtered.groupby(
    "user_id", group_keys=False
).apply(compute_gen_bbox)
```

# What is the Minimal Bounding Box (in simple terms)
Problem with general bounding box

From our plots, we already saw this:

- One far-away visit → bounding box explodes

- That does not represent where the user usually eats

So we need to ask:

“What if I only look at where the user eats most of the time?”

## Minimal bounding box = core dining region

Definition (from senior’s work):

- The minimal bounding box is the smallest geographic region that contains 80% of a user’s visited restaurants.

Key idea:

- Ignore rare outliers (travel, holidays)

- Capture habitual behavior

### How the senior computed it (algorithm)

For each user:

- Take all visited restaurant locations

- Sort them by latitude and longitude

- Slide an 80% window:
0–80%
10–90%
20–100%

- Compute bounding box area for each window

- Pick the smallest non-zero area

```python
def bbox_area(lat_min, lat_max, lon_min, lon_max):
    return (lat_max - lat_min) * (lon_max - lon_min)
```

```python
import numpy as np

def compute_min_bbox(df, coverage=0.8):
    """
    Compute minimal bounding box covering `coverage` proportion of locations
    """
    lats = np.sort(df["latitude"].values)
    lons = np.sort(df["longitude"].values)

    n = len(lats)
    window = int(np.ceil(coverage * n))

    min_area = np.inf

    for start in range(0, n - window + 1):
        lat_slice = lats[start:start + window]
        lon_slice = lons[start:start + window]

        area = bbox_area(
            lat_slice.min(),
            lat_slice.max(),
            lon_slice.min(),
            lon_slice.max()
        )

        if area > 0:
            min_area = min(min_area, area)

    return min_area if np.isfinite(min_area) else 0
```

```python
# Compute minimal bounding box per user
min_bbox_area = (
    core_filtered
    .groupby("user_id")
    .apply(compute_min_bbox)
    .rename("min_bbox_area")
)

# Merge back
core_filtered = core_filtered.merge(
    min_bbox_area,
    on="user_id",
    how="left"
)
```

Now each row has:

- bbox_area → general bounding box
- min_bbox_area → core dining region

area_ratio = general_bbox_area / minimal_bbox_area
Interpretation:

- Low ratio (~1) → very local user

- High ratio → exploratory user

```python
# Take final general bbox area per user
final_gen_bbox = (
    core_filtered
    .groupby("user_id")
    .tail(1)[["user_id", "bbox_area"]]
    .set_index("user_id")
)

area_ratio = (
    final_gen_bbox["bbox_area"] / min_bbox_area
).replace([np.inf, -np.inf], np.nan).dropna()

area_ratio.head()
```

```python
plt.figure(figsize=(7, 5))
plt.hist(area_ratio, bins=50)
plt.yscale("log")
plt.xlabel("Area Ratio (General / Minimal Bounding Box)")
plt.ylabel("User Count (log scale)")
plt.title("Distribution of User Area Ratios")
plt.show()
```

We see:

This is the empirical proof that:

- Users differ in mobility behavior

# Running KMeans Clustering(Local Expert vs Explorer)

```python
from sklearn.cluster import KMeans

X = area_ratio.values.reshape(-1, 1)
```

```python
kmeans = KMeans(n_clusters=2, random_state=42)
labels = kmeans.fit_predict(X)
```

```python
cluster_centers = kmeans.cluster_centers_.flatten()

# Smaller ratio = Local Expert
local_cluster = cluster_centers.argmin()
explorer_cluster = cluster_centers.argmax()

user_type = pd.Series(
    labels,
    index=area_ratio.index
).map({
    local_cluster: "Local Expert",
    explorer_cluster: "Explorer"
})

user_type.value_counts()
```

## Why do you see “498 Local Experts vs 2 Explorers”?

This is normal at this stage, and there are two very important reasons:
- Reason 1: I am working on a SMALL SAMPLE
I used nrows=100000. Then, filtered to restaurants, applied 5-core filtering. Only ~500 users survive. With such a small dataset: True explorers are rare, KMeans isolates extreme outliers.

So the imbalance is expected.

- Reason 2: Area ratio has extreme outliers

Look at our plot:
- Most users: area ratio ≈ 1–100
- A few users: area ratio ≈ hundreds of thousands

KMeans:

- Will ALWAYS isolate extreme points

- That’s mathematically correct

Your plot confirms:

- Explorers exist, but they are rare and extreme

That matches the senior’s observation.

```python
plt.figure(figsize=(7, 5))
plt.hist(area_ratio[user_type == "Local Expert"], bins=50, alpha=0.7, label="Local Expert")
plt.hist(area_ratio[user_type == "Explorer"], bins=50, alpha=0.7, label="Explorer")
plt.yscale("log")
plt.xlabel("Area Ratio")
plt.ylabel("User Count (log scale)")
plt.legend()
plt.title("User Mobility Clusters Based on Area Ratio")
plt.show()
```

After computing general and minimal bounding boxes, I derived an area ratio to quantify user mobility. The distribution is highly skewed, indicating heterogeneous spatial behavior. Using KMeans clustering with k=2, users naturally separate into localized and exploratory groups, aligning with the Local Expert and Explorer definitions in my senior’s work.

Due to the reduced sample size, exploratory users form a very small but distinct cluster with extremely large area ratios. This is consistent with the intuition that highly exploratory users are rare, and the proportion is expected to increase when scaling to the full dataset.

# Statistical Test 1 — Mann–Whitney U Test

What this tests (plain English)

Are area ratios of Local Experts and Explorers statistically different?

Non-parametric → perfect for skewed data.

```python
from scipy.stats import mannwhitneyu

local_ratios = area_ratio[user_type == "Local Expert"]
explorer_ratios = area_ratio[user_type == "Explorer"]

u_stat, p_value = mannwhitneyu(
    local_ratios,
    explorer_ratios,
    alternative="two-sided"
)

u_stat, p_value
```

### Mann–Whitney U Test

A Mann–Whitney U test was conducted to determine whether the area ratio
distributions of Local Experts and Explorers differ significantly.

The test is non-parametric and suitable due to the highly skewed
distribution of area ratios.

p < 0.05 → significant difference

What this means:

- A U-statistic near 0 means almost all Explorer area ratios are larger than Local Expert ratios

- p < 0.05 → reject the null hypothesis

Therefore:

- The distributions of area ratios for Local Experts and Explorers are statistically significantly different

This validates that:

- Our clustering is not arbitrary

- Area ratio meaningfully separates users

# Statistical Test 2 — Point-Biserial Correlation

What this tests

How strongly is area ratio associated with user type?

```python
from scipy.stats import pointbiserialr

binary_labels = (user_type == "Explorer").astype(int)

corr, corr_p = pointbiserialr(
    binary_labels.loc[area_ratio.index],
    area_ratio
)

corr, corr_p
```

### Point-Biserial Correlation

Point-biserial correlation was used to quantify the relationship between
user mobility class (Local Expert vs Explorer) and area ratio.

A significant positive correlation indicates that higher area ratios
are strongly associated with exploratory user behavior.

What this means:

- Correlation ≈ 0.92 → extremely strong positive relationship

- As area ratio increases → probability of being an Explorer increases

- p-value is effectively zero → not due to chance

I used point-biserial correlation to quantify the relationship between user mobility class and area ratio. The results show a very strong positive correlation, indicating that higher area ratios are strongly associated with exploratory user behavior. This supports the use of area ratio as a meaningful mobility feature.
